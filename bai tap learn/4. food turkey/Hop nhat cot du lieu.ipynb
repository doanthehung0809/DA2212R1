```python
import pandas as pd
df = pd.read_csv("E:\Gymcode\DA\DA - Datasets\FoodPrice_in_Turkey.csv")
df.head()
```

```python
## Xóa các dòng có thuộc tính ProductID trùng nhau, giữ lại bản ghi cuối cùng, giữ chỉ số ban đầu của các dòng
df = df.drop_duplicates('ProductName', keep='last')
df.head(5)
```
